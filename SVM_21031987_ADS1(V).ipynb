```python
import pandas as pd
import matplotlib.pyplot as plt

'''Reading the Excel File with Pandas'''

df_data = pd.read_excel(r'C:\Users\vmsav\Desktop\Data_vm.xlsx')
```

```python
def lineplot():
    
    '''    Lineplot function is created to plot multiple line plots with 
    x and y axes considering the columns from the dataframe uploaded 
    with the help of pandas. This plot uses the built-in functions 
    of the matplotlib.pyplot module '''

    #Using groupby function to group the data with country
    df = df_data.groupby(['country']).mean()

    #Assingning the selected countries to variables
    US = df_data[df_data.country == 'United States']
    China = df_data[df_data.country == 'China']
    India = df_data[df_data.country == 'India']
    UK = df_data[df_data.country == 'United Kingdom']
       
    plt.figure()

    #Using plot function for the line plot with x and y axis
    plt.plot(US.year,US.gdpPercap)
    plt.plot(China.year,China.gdpPercap)
    plt.plot(India.year,India.gdpPercap)
    plt.plot(UK.year,UK.gdpPercap)
    
    #labelleling the x and y axes
    plt.xlabel('Years')
    plt.ylabel('GDP Per Capita[In USD]')
        
    #legend function for giving more information regarding the plots
    plt.legend(['United States','China','India','UK'])
        
    #providing a title to our plot
    plt.title('Comparision of GDP among 4 selected countries')
    
    #Saving the plot to the local folder
    plt.savefig('C:/Users/vmsav/Desktop/Vepurimohan_21031987_ADS1/Images/Lineplot.png')
    
    plt.show()
    
    return()

#Calling the function
lineplot()

help(lineplot)
```

```python
#Grouping the data based on the continents
df_con = df_data.groupby(['continent']).median()
print(df_con)
```

```python
def barplot():
    
    '''    barplot function is used to compare the data from the 
    uploaded excel file. One axis of the plot show the different 
    continents compared among them with the population measure 
    taken on the other axes. '''
    
    #Assigning the required data into desired variables
    continents = ['Africa','America','Asia','Europe','Oceania']
    colors = ['green','yellow','red','blue','pink']
    
    plt.figure()

    #mentioning the function of the type of graph to be used
    plt.bar(continents,df_con['population'],color= colors,alpha=.5,align='center')

    #Labelling the x and y axis of the graph & providing a title
    plt.xlabel('Continents')
    plt.ylabel('Population[In Billions]')
    plt.title('Continents with highest and lowest population')
    
    #Saving the image to local folder
    plt.savefig('C:/Users/vmsav/Desktop/Vepurimohan_21031987_ADS1//Images/barplot.png')

    plt.show()
    return()

#Calling the function
barplot()

help(barplot)
```

```python
def pieplot():
    
    '''  pieplot function is created to implement a pie chart with pandas and matplotlib.pyplot
    it is used to show a parts-to-whole relationship for categorical or nominal data. 
    The slices in the pie typically represent percentages of the total.'''
    
    continents = ['Africa','America','Asia','Europe','Oceania']
    
    #mentioning the function of the type of graph to be used
    plt.pie(df_con['lifeExp'],labels=continents,autopct='%1.0f%%')
    
    #Providing the title to the graph
    plt.title('Life Expectancy among different continents')
    
    # Saving the image to a local folder
    plt.savefig('C:/Users/vmsav/Desktop/Vepurimohan_21031987_ADS1/Images/pieplot.png')
    
    plt.show()
    return()
pieplot()
help(pieplot)
```
